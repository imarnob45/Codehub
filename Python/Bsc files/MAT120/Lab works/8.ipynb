```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
#1

x = np.linspace(-2*np.pi, 2*np.pi, 1000)
a = 2
y = a * np.sin(3*x)
y1 = a * 3 * np.cos(3*x)
y2 = -a * 9 * np.sin(3*x)
y3 = -a * 27 * np.cos(3*x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = a*sin(3x)')
plt.plot(x, y1, label="y' = 3a*cos(3x)")
plt.plot(x, y2, label="y'' = -9a*sin(3x)")
plt.plot(x, y3, label="y''' = -27a*cos(3x)")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = a*sin(3x) and its Derivatives')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#2

x = sp.symbols('x')
y = sp.acos(x) * sp.exp(x)
y1 = y.diff(x)
y2 = y1.diff(x)

ans = y2 + 2 * (y1**3) - 6 * y
simp_res = sp.simplify(ans)
print(simp_res)
```

```python
#3.a

from scipy.integrate import odeint
m = 1
F = 10
mu = 0.5
def model(v, t):
    return (F - mu * v) / m
v0 = 0
t = np.linspace(0, 10, 100)
v = odeint(model, v0, t)
print(v)
```

```python
#3.b

from scipy.integrate import odeint
m = 1
F = 10
mu = 0.5
def model(v, t):
  return (F - mu * v) / m
v0 = 0
t = np.linspace(0, 10, 100)
v = odeint(model, v0, t)
plt.plot(t, v)
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.title('Velocity vs Time')
plt.show()
```

```python
#4

a = 2
def y1(x):
  return np.sqrt(4 * a * x)

def y2(x):
  return np.sqrt(4 * a * x)

x = np.linspace(0, 5, 1000)

plt.figure(figsize=(6,6))
plt.plot(x, y1(x), label='y^2 = 4ax')
plt.plot(y2(x), x, label='x^2 = 4ay')
plt.fill_between(x, y1(x), y2(x), where=(y1(x) >= y2(x)), color='gray', alpha=0.3)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Area between y^2 = 4ax and x^2 = 4ay')
plt.legend()

plt.grid(True)
plt.show()
```

```python
#5.a

from scipy.integrate import quad
def integrand(phi):
  return (2 * phi * (np.arctan(phi**2))**2) / (1 + phi**4)
ans, error = quad(integrand, 0, np.pi/2)
print(ans)
```

```python
#5.b

from scipy.integrate import quad
def integrand(x):
  return x * np.log(x)

ans, error = quad(integrand, 1, np.sqrt(np.e))
print(ans)
```

```python
#5.c

x = sp.symbols('x')
expression = ((x**2) + 3*x + 1) / ((x-1)*(x+3)*(x-2))
integral = sp.integrate(expression, x)
ans = integral.evalf()
print(ans)
```

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n - 1)
num = int(input("Enter your value: "))

ans = factorial(num)
print('Desired value: ', ans)
```
