```python
# A.

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def integrand(x):
    return np.sin(x**2)
a = -10
b = 10
x = np.linspace(a, b, 1000)
y = np.zeros_like(x)
for i in range(len(x)):
    y[i], _ = quad(integrand, a, x[i])

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Approximation of sin(x^2)')
plt.show()
```

```python
#C

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# Define the integrand
def integrand(x):
    return x**x

# Define the limits of integration
a = 0
b = 1

# Create an array of x values
x = np.linspace(a, b, 100)

# Evaluate the integral at each x value using numerical integration
y = np.zeros_like(x)
for i in range(len(x)):
    y[i], _ = integrate.quad(integrand, a, x[i])

# Plot the graph of the integral
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('∫(x^x)dx')
plt.title('Graph of the Integral ∫(x^x)dx')
plt.show()
```

```python
# D.

import numpy as np
import matplotlib.pyplot as plt

def parabola1(x):
    return x**2
def parabola2(x):
    return -x**2 + 4
x = np.linspace(-2, 2, 50)

plt.plot(x, parabola1(x), label='y = x^2')
plt.plot(x, parabola2(x), label='y = -x^2 + 4')
plt.fill_between(x, parabola1(x), parabola2(x), where=(parabola1(x) > parabola2(x)), color='gray', alpha=0.3)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Intersection of Two Parabolas')
plt.legend()
plt.grid(True)
plt.show()
```
