```python
#1

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def f(x):
    return (np.sin(x) + np.cos(x)) / np.sqrt(1 + np.sin(2*x))

def integral(x):
    result, _ = quad(f, 0, x)
    return result

x = np.linspace(-np.pi/4, np.pi/4, 1000)

y = np.array([integral(xi) for xi in x])

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = ∫(sinx+cosx)/√(1+sin2x)dx')
plt.grid(True)
plt.show()
```

```python
#2

import numpy as np
import matplotlib.pyplot as plt

def func(a, x):
    return np.sqrt(-4 + a**x)
def trapezoidal_rule(func, a, b, n):
    h = (b - a) / n
    integral = (func(a) + func(b)) / 2
    for i in range(1, n):
        integral += func(a + i * h)
    return integral * h
a = 2
x_values = np.linspace(0, 2, 100)
y_values = func(a, x_values)
area = trapezoidal_rule(lambda x: func(a, x), 0, 2, 100)
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'y^2 = -4 + {a}^x')
plt.fill_between(x_values, y_values, color='skyblue', alpha=0.3)
plt.title(f'Area between y^2 = -4 + {a}^x and y-axis: Area = {area:.4f}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
