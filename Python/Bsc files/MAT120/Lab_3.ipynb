##Plotting graphs

```python
#importing matplotlib
import matplotlib.pyplot as plt

# x & y axis values
x = (1,3,5,7,9)
y = (2,4,6,8,10)

# plotting the graph
plt.plot(x, y)

# naming axes
plt.xlabel('x - axis')
plt.ylabel('y - axis')

# title
plt.title('hello graph!')

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10,11,0.5)
y = 4*x**2

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("parabolic function")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10,11)
f = 2*x**3+ x**2 + 3*x+ 5

#plotting
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f')
plt.title("Cubic function")
plt.show()
```

```python
#lets mix these two up

#importing library modules
import matplotlib.pyplot as plt
import numpy as np

#defining variables
x = np.arange(-10,11)
y = 4*x**2
f = 2*x**3+ x**2 + 3*x+ 5

#plotting
plt.plot(x,y, label="parabola")
plt.plot(x,f, label="cubic", color="red")
plt.xlabel('x')
plt.ylabel('y,f')
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

###Plotting A Circle!

```python
import matplotlib.pyplot as plt
import numpy as np

r=int(input("Enter radius: "))
t=np.linspace(0, 2*np.pi)
x= r*np.cos(t)
y= r*np.sin(t)

plt.plot(x,y)
plt.fill_between(x,y)
plt.axis('equal')
plt.show()
```
