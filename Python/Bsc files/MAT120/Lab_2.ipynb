##Differentiation with multiple variable

```python
from sympy import*

x,y,z= symbols("x, y, z")

f = x**3 + 4*x*y - z

#performing multiple substitutions at once
fValue= f.subs([(x, 2), (y, 4), (z, 0)]) #putting values of x and y and z

df= diff(f,x)

df2 = diff(f,x,2)

##printing

print(fValue)

print(df)

print(df2)


#Introducing Lambdify Function
f = lambdify([x,y,z], f) #converting SymPy expression to a numerical expression

df = lambdify(x, df)

print(f(3,2,1))

print(df(3))
```

```python
#partial differentiation

from sympy import*

x,y,z= symbols("x, y, z")

f = x**3 + 4*x*y - z

dfx= diff(f,x)
dfy= diff(f,y)
dfz= diff(f,z)

print(dfx)
print(dfy)
print(dfz)

sum= dfx + dfy+ dfz
print(sum)
```
